```python
import imageio.v3 as imageio

def load_image(image_path):
    return imageio.imread(image_path)

image_path = "einstein.jpg"  # Change this to your image path
image = load_image(image_path)
```

```python
def check_dimensions(image):
    return image.shape

dimensions = check_dimensions(image)
print("Image Dimensions:", dimensions)
```

```python
import numpy as np

def is_integer(image):
    return np.issubdtype(image.dtype, np.integer)

integer_check = is_integer(image)
print("Is Integer Array:", integer_check)
```

```python
def convert_to_grayscale(image):
    if len(image.shape) == 3:  # Convert only if it's an RGB image
        return np.dot(image[..., :3], [0.2989, 0.5870, 0.1140]).astype(np.uint8)
    return image

grayscale_image = convert_to_grayscale(image)
```

```python
def adjust_brightness(image, value):
    return np.clip(image + value, 0, 255).astype(np.uint8)

lightened_image = adjust_brightness(grayscale_image, 50)
darkened_image = adjust_brightness(grayscale_image, -50)
```

```python
def adjust_contrast(image, factor):
    mean = np.mean(image)
    return np.clip((image - mean) * factor + mean, 0, 255).astype(np.uint8)

contrast_image = adjust_contrast(grayscale_image, 1.5)
```

```python
def crop_image(image, x_start, x_end, y_start, y_end):
    return image[y_start:y_end, x_start:x_end]

cropped_image = crop_image(grayscale_image, 50, 200, 50, 200)
```

```python
def flip_horizontally(image):
    return np.flip(image, axis=1)

flipped_image = flip_horizontally(grayscale_image)
```

```python
def rotate_90(image):
    return np.rot90(image)

rotated_image = rotate_90(grayscale_image)
```

```python
import matplotlib.pyplot as plt

def show_image(image, title="Image"):
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Display the images
show_image(image, "Original Image")
show_image(grayscale_image, "Grayscale Image")
show_image(lightened_image, "Lightened Image")
show_image(darkened_image, "Darkened Image")
show_image(contrast_image, "High Contrast Image")
show_image(cropped_image, "Cropped Image")
show_image(flipped_image, "Flipped Image")
show_image(rotated_image, "Rotated Image")
```
